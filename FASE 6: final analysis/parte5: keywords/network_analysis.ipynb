```python
import json
import pandas as pd
import math
import spacy
import numpy as np
```

```python
nlp = spacy.load("en_core_web_md")
```

```python
path = "C:\\Users\\Tiziano Pacifico\\Desktop\\TESI General\\TESI_FINAL_STAGE\\FASE 6\\DFs\\"
filename = "final_df.json"
df = pd.read_json(path+filename, lines=True)
```

```python
topics = list(df['GPT_label'].unique())
```

```python
path = "C:\\Users\\Tiziano Pacifico\\Desktop\\TESI General\\TESI_FINAL_STAGE\\FASE 6\\DFs\\"
file = "freq_words_per_topic.json"
with open(path+file, 'r') as file_json:
    # Usa json.load() per caricare il contenuto del file JSON in un dizionario
    freq_words_per_topic = json.load(file_json)
```

```python
#calcolo il numero di parole per ogni topic
n_words_per_topic = {}
for topic in topics:
    n_words_per_topic[topic] = 0
    for word, freq in freq_words_per_topic[topic].items():
        n_words_per_topic[topic] += freq
```

```python
n_different_words_per_topic = {}
for topic in topics:
    n_different_words_per_topic[topic] = len(list(freq_words_per_topic[topic].keys()))
    
```

```python
#tf
tf_per_topic = {}
for topic in topics:
    tf_per_topic[topic] = {}
    for word, freq in freq_words_per_topic[topic].items():
        tf_per_topic[topic][word] = freq_words_per_topic[topic][word]/n_words_per_topic[topic]
```

```python
#liste di parole per topic
word_list_per_topic = {}
for topic in topics:
    word_list_per_topic[topic] = set(freq_words_per_topic[topic].keys())
```

```python
#idf
n_doc = len(topics)
idf_per_topic = {}
for topic in topics:
    print(topic)
    idf_per_topic[topic] = {}
    for word, freq in freq_words_per_topic[topic].items():
        n_docs_with_word = sum(1 for word_list in word_list_per_topic.values() if word in word_list)
        idf_per_topic[topic][word] = math.log(len(topics)/n_docs_with_word+1)
```

```python
tf_idf_per_topic = {}
for topic in topics:
    tf_idf_per_topic[topic] = {}
    for word in tf_per_topic[topic].keys():
        tf_idf_per_topic[topic][word] = tf_per_topic[topic][word] * idf_per_topic[topic][word]
```

```python
similar_words_matrix = {}
for tx in topics:
    similar_words_matrix[tx] = {}
    for ty in topics:
        similar_words_matrix[tx][ty] =  word_list_per_topic[tx] & word_list_per_topic[ty]
```

```python
def calcScore(tx,ty):
    sommaTfIdf = 0
    for word in similar_words_matrix[tx][ty]:
        sommaTfIdf += (tf_idf_per_topic[tx][word] + tf_idf_per_topic[ty][word])/2
    mediaTfIdf = sommaTfIdf / len(similar_words_matrix[tx][ty])
    dimMediaTopic = (n_different_words_per_topic[tx] + n_different_words_per_topic[ty])/2
    #return (mediaTfIdf/dimMediaTopic)*len(similar_words_matrix[tx][ty])
    return (len(similar_words_matrix[tx][ty])/dimMediaTopic)
    
```

```python
similar_words_weighted_score_matrix = {}
for tx in topics:
    similar_words_weighted_score_matrix[tx] = {}
    for ty in topics:
        similar_words_weighted_score_matrix[tx][ty] = calcScore(tx,ty)
```

```python
selected_keyword_per_topic = {}
for topic, dizionario in tf_idf_per_topic.items():
    selected_keyword_per_topic[topic] = list(dict(sorted(dizionario.items(), key=lambda item: item[1], reverse=True)[0:70]).keys())
        
```

```python
weighted_embeddings_per_topic = {}
for topic in topics:
    tfidfs = []
    embeddings = []
    for word in selected_keyword_per_topic[topic]:
        tfidfs.append(tf_idf_per_topic[topic][word])
        embeddings.append(nlp(word).vector)
    weighted_embeddings_per_topic[topic] = [embedding * tfidf for embedding,tfidf in zip(embeddings,tfidfs)]
```

```python
aggregated_embeddings_per_topic = {}
for topic in topics:
    aggregated_embeddings_per_topic[topic] = np.sum(weighted_embeddings_per_topic[topic], axis = 0)
```

```python
def cosine_similarity(vec1, vec2):
    """Calcola la similarit coseno tra due vettori."""
    dot_product = np.dot(vec1, vec2)
    norm_vec1 = np.linalg.norm(vec1)
    norm_vec2 = np.linalg.norm(vec2)
    return dot_product / (norm_vec1 * norm_vec2)
```

```python
cossim_matrix = {}
for tx in topics:
    cossim_matrix[tx] = {}
    for ty in topics:
        cossim_matrix[tx][ty] = cosine_similarity(aggregated_embeddings_per_topic[tx], aggregated_embeddings_per_topic[ty])
```

```python
score = []
for tx in topics:
    for ty in topics:
        score.append(cossim_matrix[tx][ty])
```

```python
weighted_embeddings_per_topic_label = {}
for topic in topics:
    weighted_embeddings_per_topic_label[topic] = nlp(topic).vector
```

```python
cossim_matrix_label = {}
for tx in topics:
    cossim_matrix_label[tx] = {}
    for ty in topics:
        cossim_matrix_label[tx][ty] = cosine_similarity(weighted_embeddings_per_topic_label[tx],weighted_embeddings_per_topic_label[ty])
```

```python
score_ = []
for tx in topics:
    for ty in topics:
        score_.append(cossim_matrix_label[tx][ty])
```
